```python
# Storing the first three columns in X and the last column in y
X = dataset.iloc[:, :-1]  #  all columns except the last row
y = dataset.iloc[:, -1]   # last column

# Displaying the results
print("Features (X):")
print(X)
print("\nTarget (y):")
print(y)
```

```python
#Handling the missing values
from sklearn.impute import SimpleImputer
# Displaying original X
print("Original Features (X):")
print(X)


# Handling missing values in X

# Impute numerical columns with mean
numeric_cols = ['Age', 'Blood Pressure','Cholesterol Level' , 'BMI' ,  'Sleep Hours' , 'Triglyceride Level', 'Fasting Blood Sugar',  'CRP Level'  ,'Homocysteine Level']
numeric_imputer = SimpleImputer(strategy='mean')
X[numeric_cols] = numeric_imputer.fit_transform(X[numeric_cols])



# Impute categorical column with mode
categorical_col = ['Gender',  'Exercise Habits',  'Smoking',  'Family Heart Disease',  'Diabetes',   'High Blood Pressure',  'Low HDL Cholesterol',   'High LDL Cholesterol',  'Alcohol Consumption',   'Stress Level',   'Sugar Consumption']
categorical_imputer = SimpleImputer(strategy='most_frequent')
X[categorical_col] = categorical_imputer.fit_transform(X[categorical_col])


# Displaying the updated X after handling missing values
print("\nUpdated Features (X) after handling missing values:")
print(X)







```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# List of categorical columns to encode
categorical_cols = ['Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease', 'Diabetes', 'High Blood Pressure',
                    'Low HDL Cholesterol', 'High LDL Cholesterol', 'Alcohol Consumption', 'Stress Level', 'Sugar Consumption']

 # Loop through each categorical column and apply label encoding
for col in categorical_cols:
       X[col] = label_encoder.fit_transform(X[col])

# Display the updated DataFrame
print("\nUpdated Features (X) after Label Encoding:")
print(X)

y=label_encoder.fit_transform(y)
# Display the updated DataFrame
print("\nUpdated Features (y) after Label Encoding:")
print(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python

x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
x_train_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
# Building Decision Tree model
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train_scaled, y_train)
```

```python
dt_predictions = dt_model.predict(x_test_scaled)
print("Decision Tree Classification Report:")
print(classification_report(y_test, dt_predictions))
```
